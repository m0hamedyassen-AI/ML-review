# **Feature Selection**

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectKBest, chi2, f_regression , mutual_info_classif
from sklearn.feature_selection import RFE
```

```python
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
person_filter = SelectKBest(f_regression,k=3)
```

```python
x_new = person_filter.fit_transform(X,y)
```

```python
x_train ,x_test ,y_train ,y_test =train_test_split(x_new,y , test_size=.3)
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
mean_squared_error(y_pred,y_test)
```

```python
wrapp = RFE(model , n_features_to_select=3)
```

```python
x_new=wrapp.fit_transform(X,y)
```
